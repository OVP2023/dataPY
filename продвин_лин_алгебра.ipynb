Задание 1
При помощи линейных отображений:

    Уменьшить объект в два раза
    Повернуть на 130 градусов
    Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])


```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
B=np.array([[0.5,0,0],[0,0.5,0],[0,0,0.5]])  #диагональная матрица количество столбцов=3
B
```

```python
A.transpose()  #Транспонированная матрица А количество строк=3
```

```python
res=B@A.transpose()
tt=res.transpose()  # для того чтобы получить визуализацию
tt

```

```python
x = tt[:,0]
y = tt[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Повернуть на 130 градусов
import math
M=np.array([[math.cos(math.radians(130)),-math.sin(math.radians(130))],[math.sin(math.radians(130)),math.cos(math.radians(130))]])
M
```

```python
A2=[
[-100, -100],
[-100, 100],
[100, 100],
[0, 0],
[100, -100],
[-100, -100]
]
A2 = np.array(A2)
res=M@A2.transpose()
ttt=res.transpose()
```

```python
x = ttt[:,0]
y = ttt[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
#Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])
MO=[[0, 1, 0], [1, 0, 0], [0, 0, 1]]
MO = np.array(MO)
res=MO@A.transpose()
tttt=res.transpose()
```

```python
x = tttt[:,0]
y = tttt[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
AA=np.array([[1, 2], [2, 3]])
```

```python
#Спектральное разложение матрицы AA
#Получаем собственные значения и собственные вектора матрицы AA
np.linalg.eig(AA)
```

```python
 #Получаем матрицу VS из собственных векторов матрицы AA
 VS=np.array([[-0.85065081,0.52573111],[0.52573111,-0.85065081]])
 VS
```

```python
#Получаем транспонированную матрицу VS
SI=np.transpose(VS)
SI
```

```python
#Получаем диагональную матрицу DI из собственных значений матрицы AA
DI=np.array([[-0.23606798,0],[0,4.23606798]])
DI
```

```python
#Праверяем правильность результата, два значения в результирующей матрице неверны
R=VS@DI@SI
R
```

```python
#Сингулярное разложение матрицы AA
#Получение: U и V - ортогональные матрицы, элементы которых — левые и правые сингулярные вектора и сингулярные числа s
U, s, V = np.linalg.svd(AA, full_matrices=True)
print('Матрица U')
print(U)
print('Матрица V')
print(V)
print('Cингулярные числа')
print(s)
```

```python
#Формируем диагональную матрицу D из сингулярных чисел s матрицы AA
D=([[4.23606798,0],[0,0.23606798]])
print('Матрица D')
print(D)
```

```python
#Проверка правильности спектрального разложения матрицы AA
F=U@D@V
F
```
