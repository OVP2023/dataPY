```python
import numpy as np
import pandas as pd
import scipy.stats as st
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
```

```python
#Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы
headlist=['id_user','id_item','rating','timestamp']
df_rating = pd.read_csv("https://raw.githubusercontent.com/OVP2023/dataPY/main/u.data.csv",sep="\t", header=None, names=headlist)
df_rating.head()
```

```python
headlist2=['id_item','movie_title','release_date','video release_date','IMDb URL','unknown','Action','Adventure','Animation',"Children's",'Comedy','Crime','Documentary','Drama','Fantasy' ]
headlist2=headlist2+ ['Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
df_movies = pd.read_csv("/content/u.item.csv",sep="|",encoding = "ISO-8859-1", header=None, names=headlist2)
df_movies.head()
```

```python
#Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок
#ID=405
df_rating['id_user'].value_counts()
```

```python
#Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь
df_rating=df_rating[df_rating['id_user']==405]
df_rating.info()
```

Добавьте к датафрейму из прошлого задания  столбцы:
По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет,также
столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
#через join  on='id_item'

result=df_rating.merge(df_movies,on='id_item', how='left')
result=result.drop(['movie_title', 'video release_date', 'IMDb URL', 'unknown'],	axis=1)
result.head()
```

```python
result.info()
```

Для построения модели "Рекомендации оценки фильма (rating) пользователям" нужны следующие признаки:
Жанры,
(Год выхода,
Общее количество оценок,
Суммарная оценка).


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics

X=result[['Action','Adventure','Animation',"Children's",'Comedy','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']] # признаки X
y=result['rating'] # целевая переменная y

#Сформируйте X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

#Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах
model = LinearRegression()
model.fit(X_train, y_train)

# Коэффициенты линейной регрессии по каждому признаку
model.coef_
```

```python
#Оценка модели по метрикам R2, MAE,MSE,RMSE
y_pred = model.predict(X_test)
print(f'R2: {metrics.r2_score(y_test, y_pred)}')
print(f'MAE: {metrics.mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {metrics.mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {np.sqrt(metrics.mean_squared_error(y_test, y_pred))}')
```
